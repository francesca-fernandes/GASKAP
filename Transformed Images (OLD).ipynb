```python
from astropy.io import fits #looking at the fits file
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from PIL import Image #so you can open the image
```

```python
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
```

```python
image1 = Image.open('/Users/FRANNYPACK/Downloads/grind!/research data/image1.png') #the image at 1.75 * 10^4 m/s
xsize, ysize = image1.size
print(f"Image size: {ysize} x {xsize}")
plt.imshow(image1, cmap = 'gray')
plt.colorbar()
plt.xlabel('Right Ascension')
plt.ylabel('Declination')
plt.grid(color = 'white', ls = 'solid')
plt.gca().set_xticklabels(np.arange(-500, 4000, 500))
plt.gca().set_yticklabels(np.arange(-70, -76, -2))
```
